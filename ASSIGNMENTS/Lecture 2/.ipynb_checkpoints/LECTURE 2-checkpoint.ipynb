```python
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt
import seaborn as sns

# Load ARFF file with specified encoding
with open('Raisin_Dataset.arff', 'r', encoding='utf-8') as f:
    data, meta = arff.loadarff(f)

# Convert data to pandas DataFrame
df = pd.DataFrame(data)

# Decode byte strings to regular strings if needed
for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = df[column].str.decode('utf-8')
        
# checkpoint
features = df.columns[:-1]  # Exclude the 'Class' column

df
```

```python
# Set the style for the plots
sns.set(style="whitegrid")
# Step 1: Explore the distributions of each morphological feature for both classes of raisins
for feature in features:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x=feature, hue='Class', kde=True, palette='Set2')
    plt.title(f'Distribution of {feature} for Kecimen and Besni raisins')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend(title=feature, labels=['Kecimen', 'Besni'])
    plt.show()
```

```python
# Step 2: Investigate relationships between pairs of features using scatter plots
sns.pairplot(df, hue='Class', palette='Set2')
plt.suptitle('Pairplot of Morphological Features for Kecimen and Besni Raisins')
plt.show()


```

```python
# Step 3: Compare the overall distributions of features between Kecimen and Besni raisins using box plots
for feature in features:
    plt.figure(figsize=(12, 8))
    sns.boxplot(data=df, x='Class', y=feature, palette='Set2')
    plt.title(f'Boxplot of {feature} for Kecimen and Besni raisins')
    plt.xlabel('Class')
    plt.ylabel(feature)
    plt.show()
```
